```python
#stacking arrays
import numpy as np
a = np.array([[1,2],[3,5]])
b = np.array([[5,6],[7,8]])

```

```python
# stack means  add both stack(arrays) not their elements and you cant stack 2 matrix but you can do with 2 vectors
aa = np.array([1,2,3,5])
bb = np.array([5,6,7,8])
bb
```

```python
np.vstack((aa,bb))
```

```python
np.hstack((aa,bb))
```

```python
#rows becomes column like [1,2],[3,4] --->>> [1,3], [2,4]
np.column_stack((aa,bb))
```
